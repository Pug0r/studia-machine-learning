```python
import numpy as np
import pandas as pd
x = np.array([188, 181, 197, 168, 167, 187, 178, 194, 140, 176, 168, 192, 173, 142, 176]).reshape(-1, 1).reshape(15,1)
y = np.array([141, 106, 149, 59, 79, 136, 65, 136, 52, 87, 115, 140, 82, 69, 121]).reshape(-1, 1).reshape(15,1)

x = np.asmatrix(np.c_[np.ones((15,1)),x])

I = np.identity(2)
alphas = np.arange(0, 1, 0.05)
results = []

for alpha in alphas:
    w = np.linalg.inv(x.T*x + alpha * I)*x.T*y
    y_pred = x * w
    mse = np.mean(np.square(y - y_pred))
    w_flat = np.asarray(w).flatten()
    results.append([w_flat[0], w_flat[1], mse])

comparison_df = pd.DataFrame(results, columns=['Intercept', 'Slope', 'MSE'], index=alphas)
print(comparison_df)
```

## 2. Implement based on the Ridge regression example, the Lasso regression.

Please implement the SGD method and compare the results with the sklearn Lasso regression results. 

```python
def sgd(X, y, alpha, lr=0.001, epochs=1000):
    n = X.shape[1]
    w = np.zeros(n) #  strating weight are zeros
    
    for _ in range(epochs):
        for i in range(len(y)):  
            y_hat = np.dot(X[i], w)
            
            error = y_hat - y[i]
            
            gradient = X[i] * error

            l1_penalty = alpha * np.sign(w)
            l1_penalty[0] = 0  # intercept
            
            w = w - lr * (gradient + l1_penalty)
            
    return w
```

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

x = np.array([188, 181, 197, 168, 167, 187, 178, 194, 140, 176, 168, 192, 173, 142, 176]).reshape(-1, 1).reshape(15,1)
y = np.array([141, 106, 149, 59, 79, 136, 65, 136, 52, 87, 115, 140, 82, 69, 121]).reshape(-1, 1).reshape(15,1)
alpha = 0.1 

# need to scale data!! otherwise you get nans everywhere
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x = np.c_[np.ones(len(x_scaled)), x_scaled]

lasso_regressor = Lasso(alpha=alpha)
lasso_regressor.fit(np.asarray(x), np.asarray(y))
print(lasso_regressor.coef_, lasso_regressor.intercept_)

w_sgd = sgd(np.asarray(x), y.flatten(), alpha)
print("SGD Coefficients:", w_sgd)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris_data = load_iris()
iris_df = pd.DataFrame(iris_data.data,columns=iris_data.feature_names)
iris_df['target'] = iris_data.target
print(iris_df.head().columns)

x = iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']].values
y = iris_df['target'].values.reshape(-1, 1)

dataset_size = np.size(x)

mean_x, mean_y = np.mean(x), np.mean(y)

SS_xy = np.sum(y * x) - dataset_size * mean_y * mean_x
SS_xx = np.sum(x * x) - dataset_size * mean_x * mean_x

a = SS_xy / SS_xx
b = mean_y - a * mean_x

y_pred = a * x + b
df = pd.DataFrame(y_pred)
df
```
